```python
# Initial imports
import pandas as pd
import hvplot.pandas
from path import Path
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

### Deliverable 1: Preprocessing the Data for PCA

```python
# Load the crypto_data.csv dataset.
# YOUR CODE HERE
file_path = Path('crypto_data.csv')
crypto_df = pd.read_csv(file_path, index_col=0)
crypto_df.head()
```

```python
# Keep all the cryptocurrencies that are being traded.
# YOUR CODE HERE
trading_df = crypto_df.loc[crypto_df["IsTrading"] == True]
trading_df.head(10)
```

```python
# Keep all the cryptocurrencies that have a working algorithm.
# YOUR CODE HERE
trading_df = trading_df.dropna(subset=['Algorithm'])
trading_df.head(10)
```

```python
# Remove the "IsTrading" column. 
# YOUR CODE HERE
new_trading_df = trading_df.drop(["IsTrading"], axis=1)
new_trading_df.head(10)
```

```python
# Remove rows that have at least 1 null value.
# YOUR CODE HERE
nonull_trading_df = new_trading_df.dropna()
nonull_trading_df
```

```python
# Keep the rows where coins are mined.
# YOUR CODE HERE
mined_coins_df = nonull_trading_df.loc[nonull_trading_df["TotalCoinsMined"]>0]
mined_coins_df
```

```python
# Create a new DataFrame that holds only the cryptocurrencies names.
# YOUR CODE HERE
names_df = pd.DataFrame(mined_coins_df['CoinName'])
names_df
```

```python
# Drop the 'CoinName' column since it's not going to be used on the clustering algorithm.
# YOUR CODE HERE
new_crypto_df = mined_coins_df.drop(["CoinName"], axis=1)
new_crypto_df
```

```python
# Use get_dummies() to create variables for text features.
# YOUR CODE HERE
X = pd.get_dummies(new_crypto_df, columns=["Algorithm", "ProofType"])
X
```

```python
# Standardize the data with StandardScaler().
# YOUR CODE HERE
crypto_scaled = StandardScaler().fit_transform(X)
print(crypto_scaled[0:5])
```

### Deliverable 2: Reducing Data Dimensions Using PCA

```python
# Using PCA to reduce dimension to three principal components.
# YOUR CODE HERE
pca = PCA(n_components=3)
crypto_pca = pca.fit_transform(crypto_scaled)
crypto_pca
```

```python
# Create a DataFrame with the three principal components.
# YOUR CODE HERE
pcs_df = pd.DataFrame(data=crypto_pca,
                          columns=['PC 1', 'PC 2', 'PC 3'], index = new_crypto_df.index)
pcs_df
```

### Deliverable 3: Clustering Crytocurrencies Using K-Means

#### Finding the Best Value for `k` Using the Elbow Curve

```python
# Create an elbow curve to find the best value for K.
# YOUR CODE HERE
# Store values of K to plot
inertia = []
k = list(range(1, 11))

# Looking for the best K
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)
    
# Define a DataFrame to plot the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", title="Elbow Curve", xticks=k)
```

Running K-Means with `k=4`

```python
# Initialize the K-Means model.
# YOUR CODE HERE
model = KMeans(n_clusters=4, random_state=5)
model

# Fit the model
# YOUR CODE HERE
model.fit(pcs_df)

# Predict clusters
# YOUR CODE HERE
predictions = model.predict(pcs_df)
print(predictions)
```

```python
# Create a new DataFrame including predicted clusters and cryptocurrencies features.
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
# YOUR CODE HERE
clustered_df = pd.concat([new_crypto_df, pcs_df], axis = 1)

#  Add a new column, "CoinName" to the clustered_df DataFrame that holds the names of the cryptocurrencies. 
# YOUR CODE HERE
clustered_df = pd.concat([clustered_df, names_df], axis = 1)
#  Add a new column, "Class" to the clustered_df DataFrame that holds the predictions.
# YOUR CODE HERE
clustered_df["class"] = model.labels_
# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```

### Deliverable 4: Visualizing Cryptocurrencies Results

#### 3D-Scatter with Clusters

```python
# Creating a 3D-Scatter with the PCA data and the clusters
# YOUR CODE HERE
fig = px.scatter_3d(
    clustered_df,
    x="PC 1",
    y="PC 2",
    z="PC 3",
    color="class",
    symbol="class",
    hover_name="CoinName", 
    hover_data=["Algorithm"],
    width=800,
)
fig.update_layout(legend=dict(x=0, y=1))
fig.show()
```

```python
# Create a table with tradable cryptocurrencies.
# YOUR CODE HERE
clusterd1_df = clustered_df[["CoinName", "Algorithm", "ProofType", "TotalCoinSupply", "TotalCoinsMined", "class"]]
clusterd1_df.hvplot.table(sortable=True, selectable=True)
```

```python
# Print the total number of tradable cryptocurrencies.
# YOUR CODE HERE
print(f"There are {len(clusterd1_df.index)} tradable cryptocurrencies")
```

```python
# Scaling data to create the scatter plot with tradable cryptocurrencies.
# YOUR CODE HERE
mms = MinMaxScaler()
clustered_df_scaled = mms.fit_transform(clusterd1_df[['TotalCoinSupply','TotalCoinsMined']])
clustered_df_scaled
```

```python
# Create a new DataFrame that has the scaled data with the clustered_df DataFrame index.
# YOUR CODE HERE
plot_df=pd.DataFrame(
data=clustered_df_scaled, columns=['TotalCoinSupply','TotalCoinsMined'], index = clustered_df.index)
plot_df

# Add the "CoinName" column from the clustered_df DataFrame to the new DataFrame.
# YOUR CODE HERE
plot_df["CoinName"] = clusterd1_df["CoinName"]

# Add the "Class" column from the clustered_df DataFrame to the new DataFrame. 
# YOUR CODE HERE
plot_df["Class"] = clusterd1_df["class"]

plot_df.head(10)
```

```python
# Create a hvplot.scatter plot using x="TotalCoinsMined" and y="TotalCoinSupply".
# YOUR CODE HERE
plot_df.hvplot.scatter(x="TotalCoinsMined", y="TotalCoinSupply", by="Class", hover_cols=["CoinName"])
```
